## <u>***Theory Questions</u>:-***

**Q1 What is Python, and why is it popular?**
- Python is a high-level, interpreted, general-purpose programming language known for its simple syntax.
- It is popular because:
    - Easy to learn & read even for beginners to coding.
    - Huge libraries.
    - Widely used in Data Science, AI, ML
    - Cross-platform support.
    - Large community support.

**Q2 What is an interpreter in Python?**
- An interpreter is a program that executes Python code line-by-line, converting it to machine code during runtime.

**Q3 What are pre-defined keywords in Python?**
- Words that have specific, fixed meanings Python.
- Can't be used as variable names, function names.
    - Example: if, for, while, class, break, etc.

**Q4 Can keywords be used as variable names?**
- No, keywords cannot be used as variable names because they are reserved by Python syntax.

**Q5 What is mutability in Python?**
- Mutability means whether an object’s value can be changed after creation.
- Example: lists, dictionaries, sets.

**Q6 Why are lists mutable, but tuples are immutable?**
- Lists are mutable in Python because they are designed to allow modifications after creation, such as adding, removing, or changing elements, using methods like append(), remove() etc.
-  Tuples are immutable, i.e their contents cannot be altered once defined as they are fixed, unchangeable collection of items.
    - Example: tup = (1, 2, 3, 4)
                tup[2] = 99  {Trying to change the third element will give error}

**Q7 What is the difference between “==” and “is” operators in Python?**
- '==' compares value
    - Example: [1,2] == [1,2]  {True}
- 'is' Compares memory location
    - Example: [1,2] is [1,2]  {False}

**Q8 What are logical operators in Python?**
- Logical operators are used to combine and evaluate boolean expressions for truth value. Example: and, or, not
    - Example: 30 > 3 and 30 > 10 is True

**Q9 What is type casting in Python?**
- Converting one data type into another.
    - Example: int("3"), float(3)

**Q10 What is the difference between implicit and explicit type casting?**
- Implicit: Python automatically converts the type.
    - Example: 9 + 3.5 --> 12.5

- Explicit: Programmer converts the type manually.
    - Example: int("3")

**Q11 What is the purpose of conditional statements in Python?**
- Used for decision making and executing different code blocks based on conditions.
    - Example: if, else, elif

**Q12 How does the elif statement work?**
- 'elif' checks the next condition only if all the previous conditions are false.

**Q13 What is the difference between for and while loops?**
- For loop: Used when the number of iterations is known in advance, automatically handles initialization, condition checking, and iteration.
- While loop: Runs until a condition becomes false, used when the number of iterations is unknown.

**Q14 Describe a scenario where a while loop is more suitable than a for loop.**
- A while loop is more suitable than a for loop in when the number of iterations is unknown.
    - Example: A password check: the program repeatedly prompts the user to enter a password until the correct one is entered.

```python
# Q14 Write a program to reverse a string using a while loop.
s = input("Enter a string: ")
rev = ""
i = -1

while i >= -len(s):
    rev += s[i]
    i -= 1

print("Reversed string:", rev)
```

```python
# Q15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.
number = int(input("Enter a number: "))
result = 1
counter = number
while counter > 0:
    result *= counter
    counter -= 1
print(f"The factorial of {number} is {result}.")
```
